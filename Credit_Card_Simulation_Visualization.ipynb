```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from matplotlib import rcParams
import seaborn as sns
#sns.set_style("white") 
plt.style.use('fivethirtyeight')
# create a dataset
height = [99.85,100,98.18,100,100,100,100]

bars = ('RF', 'LR', 'KNN', 'DT', 'NB', 'MLPC', 'SVM') 
x_pos = np.arange(len(bars))

#patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o", "O" ]

patterns = [ "-", ".", "*","x", "o","|" , "/"]

fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(len(height)):
    ax.bar(x_pos[i], height[i], color='grey', edgecolor='black', hatch=patterns[i])

rects = ax.patches

labels = [str(height[i]) for i in range(len(height))]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom')
# Create names on the x-axis
plt.xticks(x_pos, bars)

ax = plt.axes()
#ax.set_xlim((10, 80))
ax.set_ylim((0, 100))

#plt.legend(fontsize=14);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#ax.set_yticklabels([])
#ax.set_xlim((0.65, 1.0))
plt.xlabel('Classifier Model', fontsize=16,fontweight='bold')
plt.ylabel('Accuracy (%)', fontsize=16,fontweight='bold')
rcParams.update({'figure.autolayout': True})

#plt.grid(linestyle='--', linewidth=0.5)
#plt.grid(True)
#plt.show()
plt.savefig('case2_accuracy')
```

# Comparison for with and without undersampling

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['Accuracy', 'Precision', 'Recall', 'Sensitivity', 'Specificity']
undersampling = [99.85,93.82,100,100,99.85]
non_undersampling = [98.1, 100, 13.9, 13.89, 100]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, undersampling, width, label='With Undersampling')
rects2 = ax.bar(x + width/2, non_undersampling, width, label='Without Undersampling')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Value (%)')
#ax.set_title('Comparison of Undersampling in RF Model')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)

for p in ax.patches[0:]:
    h = p.get_height()
    x = p.get_x()+p.get_width()/2.
    if h != 0:
        ax.annotate("%g" % p.get_height(), xy=(x,h), xytext=(0,4), rotation=90, 
                   textcoords="offset points", ha="center", va="bottom")

ax.set_xlim(-0.5, None)
ax.margins(y=0)
#ax.legend(loc="lower center", bbox_to_anchor=(0,1.02,1,0.08), borderaxespad=0, mode="expand")
    
fig.tight_layout()

#plt.show()
plt.savefig('Comparison_Undersampling1')
```
